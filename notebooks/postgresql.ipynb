## Cleaning Before Postgresql

```python
# uuid .. is the product id whould be uniqe
df['uuid'].duplicated().any()

# insights .. there is no duplicate uuid
```

```python
# cheak null 

df.isna().sum()
```

```python
df = df.where(df.notna(), None)
# “Keep all valid values as they are, and replace all missing values with None.”
```

```python
# type cheking: 

df.info()
```

```python
NUMERIC_COLS = ["selling_price", "mrp", "star_rating"]

for col in NUMERIC_COLS:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df.info()
```

```python
df.rename(columns={"product_detials": "product_details"}, inplace= True)
df.rename(columns={'uuid': 'product_id'}, inplace=True)
df.rename(columns={'url': 'product_url'}, inplace=True)
df.rename(columns={'images': 'image_url'}, inplace=True)
```

```python
df.info()
```

```python
# not more cleaning just simple strippinng

TEXT_COLS = ["title", "brand", "category", "colour"]

for col in TEXT_COLS:
    df[col] = df[col].astype(str).str.strip()
```

```python
df.info()
```

```python
df.head(2)
```

```python
# dropping unwanted columns: 

# date_stored, meta_data, set_product_price

df.drop(columns=['date_stored', 'meta_data', 'set_product_price'], inplace= True)
```

```python
df.info()
```

## storing data into postgresql

```python
# at fist time i am storing only 100 rows
```

```python
df_mini=df[:100]
```

```python
df_mini.info()
```

```python
import psycopg2
from psycopg2.extras import execute_values


conn = psycopg2.connect(
    host="127.0.0.1",
    port=5432,
    dbname="ecommerce_db",
    user="postgres",      # or admin
    password="postgres"
)

cursor = conn.cursor()

```

```python
CREATE_TABLE_SQL = """
CREATE TABLE IF NOT EXISTS products (
    -- Primary Key
    product_id TEXT PRIMARY KEY,

    -- UI
    image_url TEXT,
    product_url TEXT,

    -- Core Search Text (RAW)
    title TEXT,
    product_details TEXT,

    -- Filters (Categorical)
    brand TEXT,
    category TEXT,
    colour TEXT,
    size TEXT,
    competitor TEXT,

    -- Filters (Numeric)
    selling_price FLOAT,
    mrp FLOAT,
    star_rating FLOAT
);
"""

cursor.execute(CREATE_TABLE_SQL)

conn.commit()
print("✓ Table `products` is ready")
```

```python
df_mini.head(1)
```

```python
df_mini=df_mini[
    [
        "product_id", 
        "image_url",
        "product_url", 
        "title", 
        "product_details", 
        "brand",
        "category", 
        "colour",
        "size", 
        "competitor", 
        "selling_price",
        "mrp", 
        "star_rating"

        
    ]
]
```

```python
insert_query = """
    INSERT INTO products (
        "product_id", 
        "image_url",
        "product_url", 
        "title", 
        "product_details", 
        "brand",
        "category", 
        "colour",
        "size", 
        "competitor", 
        "selling_price",
        "mrp", 
        "star_rating"
    )
    VALUES %s;
    """
```

```python
records = df_mini.to_records(index=False).tolist()
execute_values(cursor, insert_query, records)
```

```python
conn.commit()
print(f"✓ Inserted {len(records)} rows into products")
```
